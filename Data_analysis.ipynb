# Data visualization

```python
no_need_to_visualized = ['AGE', 'RANDID']
binary_values = ['CURSMOKE', 'SEX', 'DIABETES', 'CURSMOKE']

for col_names in data.columns:
    if col_names in no_need_to_visualized:
        continue
    elif col_names in binary_values:
        
        ## TODO
        sex_data = data.loc[(data.PERIOD == 1), ['AGE', 'CURSMOKE'] ]
sex_data.groupby('AGE').count().plot()
    else:
        data.plot(x='AGE', y=col_names, style='o')
```

```python
sex_data = data.loc[(data.PERIOD == 1), ['AGE', 'CURSMOKE'] ]
ax = sex_data[sex_data.CURSMOKE == 1].groupby('AGE').count().plot()
sex_data.groupby('AGE').count().plot(ax=ax, title="curent smoker VS person", label="x")
ax.legend(["Current smoker", "Total population"])
#sex_data.groupby('CURSMOKE').count().plot(ax=ax)
```
